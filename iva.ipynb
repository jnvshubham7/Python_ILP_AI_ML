```python
import numpy as np
from sklearn.decomposition import FastICA
import matplotlib.pyplot as plt

# Generate synthetic mixed signals
np.random.seed(42)
n_samples = 2000
time = np.linspace(0, 8, n_samples)

# Generate two independent signals
s1 = np.sin(2 * time)  # Signal 1: sine wave
s2 = np.sign(np.sin(3 * time))  # Signal 2: square wave
S = np.c_[s1, s2]  # Combine signals

# Mix signals to simulate observed data
A = np.array([[1, 1], [0.5, 2]])  # Mixing matrix
X = np.dot(S, A.T)  # Mixed signals

# Apply ICA
ica = FastICA(n_components=2, random_state=42)
S_ = ica.fit_transform(X)  # Recovered signals
A_ = ica.mixing_  # Estimated mixing matrix

# Plot original, mixed, and recovered signals
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.title("Original Signals")
plt.plot(S)
plt.subplot(3, 1, 2)
plt.title("Mixed Signals")
plt.plot(X)
plt.subplot(3, 1, 3)
plt.title("Recovered Signals (ICA)")
plt.plot(S_)

plt.tight_layout()
plt.show()
```
