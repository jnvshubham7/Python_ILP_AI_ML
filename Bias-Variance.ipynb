```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.linear_model import LinearRegression
from sklearn.gaussian_process import GaussianProcessRegressor

# Create a synthetic dataset
X, y = make_moons(n_samples=100, noise=0.2)

# Implement a simple linear regression model with high bias
lr_model = LinearRegression().fit(X, y)
lr_bias = lr_model.predict(X)

# Implement a Gaussian Process model with high variance
gp_model = GaussianProcessRegressor().fit(X, y)
gp_variance = gp_model.predict(X, return_std=True)[1]

# Plot the bias-variance tradeoff
plt.plot(X[:, 0], y, 'o', label='Data')
plt.plot(X[:, 0], lr_bias, label='Linear Regression (High Bias)')
plt.plot(X[:, 0], gp_variance, label='Gaussian Process (High Variance)')
plt.legend()
plt.show()
```
